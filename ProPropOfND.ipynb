```python
import csv

with open("StudentsPerformance.csv") as f:
  reader = csv.reader(f)
  data_list = list(reader)

data_list.pop(0)
math_data = []
read_data = []
write_data = []
data = []

for i in data_list:
  math_data.append(float((i[5])))
  read_data.append(float((i[6])))
  write_data.append(float((i[7])))
```

```python
import statistics
import plotly.figure_factory as ff
import plotly.graph_objects as go
```

```python
def math():

  mean = sum(math_data)/len(math_data)
  median = statistics.median(math_data)
  mode = statistics.mode(math_data)
  std_dev = statistics.stdev(math_data)

  std_dev_start_1, std_dev_end_1 = mean-std_dev, mean+std_dev
  std_dev_start_2, std_dev_end_2 = mean-(2*std_dev), mean+(2*std_dev)
  std_dev_start_3, std_dev_end_3 = mean-(3*std_dev), mean+(3*std_dev)

  fig = ff.create_distplot(
                          [math_data],
                          ["Math Score"],
                         show_hist = False
  )

  fig.add_trace(go.Scatter(x=[mean,mean], y=[0,0.17], mode="lines", name="mean"))

  fig.add_trace(go.Scatter(x=[std_dev_start_1,std_dev_start_1], y=[0,0.17], mode="lines", name="Standard Deviation 1"))
  fig.add_trace(go.Scatter(x=[std_dev_end_1,std_dev_end_1], y=[0,0.17], mode="lines", name="Standard Deviation 1"))

  fig.add_trace(go.Scatter(x=[std_dev_start_2,std_dev_start_2], y=[0,0.17], mode="lines", name="Standard Deviation 2"))
  fig.add_trace(go.Scatter(x=[std_dev_end_2,std_dev_end_2], y=[0,0.17], mode="lines", name="Standard Deviation 2"))

  fig.add_trace(go.Scatter(x=[std_dev_start_3,std_dev_start_3], y=[0,0.17], mode="lines", name="Standard Deviation 3"))
  fig.add_trace(go.Scatter(x=[std_dev_end_3,std_dev_end_3], y=[0,0.17], mode="lines", name="Standard Deviation 3"))

  fig.show()



  list_of_data_std_1 = [result for result in math_data if result > std_dev_start_1 and result < std_dev_end_1]
  list_of_data_std_2 = [result for result in math_data if result > std_dev_start_2 and result < std_dev_end_2]
  list_of_data_std_3 = [result for result in math_data if result > std_dev_start_3 and result < std_dev_end_3]

  print("Mean -> ", mean)
  print("Median -> ", median)
  print("Mode -> ", mode)
  print("Standard Deviation -> ", std_dev)
  print("{}% of data lies within 1st standard deviation".format(len(list_of_data_std_1)*100.0/len(math_data)))
  print("{}% of data lies within 2nd standard deviation".format(len(list_of_data_std_2)*100.0/len(math_data)))
  print("{}% of data lies within 3rd standard deviation".format(len(list_of_data_std_3)*100.0/len(math_data)))
```
